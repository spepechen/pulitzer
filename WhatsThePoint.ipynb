## Gender Ratio (each year)

http://chrisalbon.com/python/pandas_create_column_with_loop.html

```python
working_df = df2['year'].value_counts().to_frame('total')
working_df
```

```python
years = list(working_df.index)
len(years)
```

```python
years
```

```python
male = []
for year in years: 
    male.append(df2[df2['year'] == year]['gender_add'].value_counts().male)
```

```python
male
```

```python
len(male)
```

Some years have zero female winner, so I need to deal with female col separtely. 

```python
female = []
for year in years: 
    if df2[df2['year'] == year]['gender_add'].value_counts().to_frame().size < 2: #male only year 
        female.append(0) 
    else:
        female.append(df2[df2['year'] == year]['gender_add'].value_counts().female)
```

```python
female
```

```python
len(female)
```

```python
working_df['male'] = male 
working_df['female'] = female
```

```python
working_df
```

```python
new_df = working_df.reset_index(level=0, inplace=False)
new_df.rename(columns={'index':'year'}, inplace=True)

new_df = new_df.sort(['year'], ascending=[True])
new_df = new_df.reset_index(drop=True)
new_df
```

http://stackoverflow.com/questions/15201386/how-to-plot-multiple-horizontal-bars-in-one-chart-with-matplotlib

```python
plt.style.use('fivethirtyeight')

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [10.0, 30.0]
print "Current size:", fig_size
 
# Set figure width to 8 and height to 30
fig_size[0] = 10
fig_size[1] = 30
plt.rcParams["figure.figsize"] = fig_size

ind = np.arange(len(new_df))
width = 0.4

fig, ax = plt.subplots()
ax.barh(ind, new_df.male, width, color='lightblue', label='M')
ax.barh(ind + width, new_df.female, width, color='pink', label='F')

ax.set(yticks=ind + width, yticklabels= new_df.year, ylim=[2*width - 1, len(new_df)])
ax.legend()

plt.show()
```

Notice: For some years there is no journalistic Pulitzer award like 1919.

## Gender Ratio (every decade)

http://stackoverflow.com/questions/17764619/pandas-dataframe-group-year-index-by-decade

```python
decade_df = new_df.groupby((new_df['year']//10)*10).sum()
decade_df
```

1910s = 1910, 1911,....,1999

2010s = 2010 ~ 2019

```python
decade_df.drop('year', 1, inplace=True)
decade_df
```

Combine 1910 and 1920 as one row, for 1910s only includes 1917 and 1918.  

```python
decade_df.ix[1920]
```

```python
# manually assign sum of two 
decade_df.ix[1920].total = 27
decade_df.ix[1920].male = 26
decade_df.ix[1920].female = 1

# delete 1910
decade_df.drop(1910, inplace=True)
decade_df
```

http://stackoverflow.com/questions/16353729/pandas-how-to-use-apply-function-to-multiple-columns

```python
def percentage(row):
    return float(row['female']) / row['total']

decade_df['ratio'] = decade_df.apply(percentage, axis=1) #axis = 1 ---> row
```

```python
decade_df
```

```python
plot_df = decade_df.copy().drop(decade_df.columns[[0,1,2]], axis=1, inplace=False)
plot_df
```

```python

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 5.0]
print "Current size:", fig_size

# Set figure width to 8 and height to 5
fig_size[0] = 8
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size

plot_df.plot()
```
