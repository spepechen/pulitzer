## Line Chart

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 5.0]
print "Current size:", fig_size

# Set new figure width to 8 and height to 5
fig_size[0] = 8
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.style.use('ggplot')

ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
plt.style.use('fivethirtyeight')

ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(range(10))

prop = fm.FontProperties(fname='/Users/Spepe/anaconda/lib/python2.7/site-packages/matplotlib/mpl-data/fonts/Lato2OFL/Lato-Regular.ttf')
ax.set_title('Jan', fontproperties=prop, size=32)

plt.show()
```
