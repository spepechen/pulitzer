```python
%matplotlib inline

import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pulitzer_winner_refined - output.csv")  #individual awards 
```

```python
df
```

```python
org_df = df[df['organizational'] == 'yes']
org_df 
```

```python
org_df['publication'].value_counts()[:10].plot(kind = 'bar', title="Top 10 organizations that win most Pulitizer (organization-wise)")

# including Public Service, and those awarded to non-specified staff
```

```python
indie_df = df[df['organizational'] == 'no']
indie_df
```

```python
indie_df['publication'].value_counts()[:10].plot(kind = 'bar', title="Top 10 organizations that win most Pulitizer (person-wise)", color = 'skyblue')
```

```python
indie_df['gender'].value_counts().plot(kind = 'bar', color = 'yellow', title = 'Gender of Personal Pulitzer Awards')
```
